```python
#Aca haremos los ejercicios del curso
import numpy as np
url = 'https://gist.githubusercontent.com/ahcamachod/9be09de793dc3bf1e6c3d98eb4e5b1ef/raw/21b85572693200040e11284ef6dcfc3457ec8e11/citrus.csv'
datos = np.loadtxt(url, delimiter=',', usecols=np.arange(1, 6, 1), skiprows=1)
```

```python
datos
```

```python
#El diámetro está en la columna cero y el peso en la columna uno.
#Los datos de las naranjas van hasta la fila 4999 y los datos de las toronjas comienzan en la fila 5000 del archivo.

diametro_naranja = datos[:5000, 0]
diametro_toronja = datos[5000:, 0]
peso_naranja = datos[:5000, 1]
peso_toronja = datos[5000:, 1]
```

```python
import matplotlib.pyplot as plt

plt.plot(diametro_naranja, peso_naranja)
plt.plot(diametro_toronja, peso_toronja)
plt.legend(['Naranja','Toronja'])
```

```python
import matplotlib.pyplot as plt
```

```python
Y = peso_naranja
X = diametro_naranja
n = np.size(X)

a = (n*np.sum(X*Y) - np.sum(X)*np.sum(Y))/(n*np.sum(X**2)-np.sum(X)**2)
b = np.mean(Y) - a*np.mean(X)
```

```python
plt.plot(X,Y)
plt.plot(8,a*8+b,'*r')
```

```python
Y_toronja = peso_toronja
X_toronja = diametro_toronja
n_toronja = np.size(X_toronja)

a_t = (n_toronja*np.sum(X_toronja*Y_toronja) - np.sum(X_toronja)*np.sum(Y_toronja))/(n_toronja*np.sum(X_toronja**2)-np.sum(X_toronja)**2)
b_t = np.mean(Y_toronja) - a_t*np.mean(X_toronja)
```

```python
plt.plot(X_toronja,Y_toronja)
plt.plot(12,a_t*12+b_t,'*r')
```
